```python
import pandas as pd
```

```python
df = pd.read_csv('db/title.episode.tsv', sep='\t', na_values='\\N')
```

```python
df.info()
```

```python
# remover episódios sem temporada ou número de episódio definidos
df = df.dropna(subset=['seasonNumber', 'episodeNumber'])

df['seasonNumber'] = df['seasonNumber'].astype(int)
df['episodeNumber'] = df['episodeNumber'].astype(int)
```

```python
df.info()
```

```python
df.to_parquet('db/episodes.parquet')
```
